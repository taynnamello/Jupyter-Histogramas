Histograma com densidade e seaborn 

```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
# Carregamento da base de dados
base = pd.read_csv('C:\\Users\\taynna.silva\\Documents\\DS\\13.Prática em Python\\dados\\chicken.csv')
base.head()
```

```python
# Criação de novo dataframe agrupando o atributo 'feed'
agrupado = base.groupby(['feed'])['weight'].sum()
agrupado
```

```python
# Novo dataframe somente para testar os filtros do pandas
teste = base.loc[base['feed'] == 'horsebean']
teste
```

```python
# Histograma considerando somente o valor 'horsebean'
srn.distplot(base.loc[base['feed'] == 'horsebean'].weight, hist = False).set_title('horsebean')
```
