Gráfico de Dispersão com Seaborn

```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
# Carregamento da base de dados
base = pd.read_csv('C:\\Users\\taynna.silva\\Documents\\DS\\13.Prática em Python\\dados\\co2.csv')
base.head()
```

```python
# Gráfico de dispersão utilizando os atributos conc e uptake. agrupado pelo type
srn.scatterplot(base.conc, base.uptake, hue = base.Type)
```

```python
# Seleção de registros especifícos da base de dados (Quebec e Mississipi)
q = base.loc[base['Type']== 'Quebec']
m = base.loc[base['Type']== 'Mississipi']
```

```python
# Subgráfico (1 linha e duas colunas) mostrando gráficos sobre cada região
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(q.conc, q.uptake).set_title('Quebec')
plt.subplot(1,2,2)
srn.scatterplot(m.conc, m.uptake).set_title('Mississipi')
plt.tight_layout()
```

```python
# Refrigerado e não refrigerado
ch = base.loc[base['Treatment']== 'chilled']
nc = base.loc[base['Treatment']== 'nonchilled']
```

```python
# Gráfico somente com chilled e nonchilled
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(ch.conc, ch.uptake).set_title('Chilled')
plt.subplot(1,2,2)
srn.scatterplot(nc.conc, nc.uptake).set_title('Nonchilled')
plt.tight_layout()
```
